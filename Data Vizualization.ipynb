## We can Chart Data with Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# here are some values to be plotted
x = np.arange(5)
y = np.random.rand(5)+3 # 5 random numbers (just over 3)
y
```

```python
# we can plot the data on a chart
plt.bar(x,y)
plt.title('Vertical Bars')
# plt.barh(x,y)
```

```python
# we can plot more than one chart within a subplot figure
plt.subplots(2) # we have a 2x2 grid for plotting charts
plt.subplot(211) # choose the top left plot of the grid
plt.bar(x,y)
plt.subplot(212) # choose the bottom left plot of the grid
plt.barh(x,y)
plt.subplot(221) # choose the top left plot of the grid
plt.bar(x,y)
plt.subplot(222) # choose the bottom left plot of the grid
plt.barh(x,y)
```

### Line Charts

```python
# Numpy has featires to generate common mathematical series
x = np.linspace(0, 6, 12) # this produces a linear progression
y = np.power(x, 2) # square each member of x
```

```python
# we can plot a line of x against x*2 (x squared)
plt.axis([0,7,0,42]) # x and y axese
plt.xlabel('linear progression')
plt.ylabel('squares')
plt.title('Line Chart')
plt.plot(x,y) # defaults to a line chart
```
